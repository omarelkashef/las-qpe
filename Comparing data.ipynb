```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1=np.load("results_10000_1.0_vqe.npy",allow_pickle=True).item()
```

```python
array2=np.load("results_1024_1.0_vqe.npy",allow_pickle=True).item()
```
